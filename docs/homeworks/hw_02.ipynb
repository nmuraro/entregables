La mayoría de los pasajeros embarcaron en Southampton (S), seguido por Cherbourg (C) y Queenstown (Q). La supervivencia varía según el puerto, posiblemente porque hay correlación con la clase del pasajero y su ubicación en el barco.



```python
import pandas as pd
import numpy as np

# Cargar datos desde tu repositorio GitHub
url_train = "https://raw.githubusercontent.com/nmuraro/entregables/main/docs/data/titanic/train.csv"
url_test = "https://raw.githubusercontent.com/nmuraro/entregables/main/docs/data/titanic/test.csv"

df_train = pd.read_csv(url_train)
df_test = pd.read_csv(url_test)

# Guardamos el PassengerId para usar en la predicción final
passenger_ids_test = df_test["PassengerId"]

# Combinamos train y test para hacer preprocesamiento conjunto
df_train["is_train"] = 1
df_test["is_train"] = 0
df_test["Survived"] = np.nan  # añadir columna faltante para igualar columnas

df_all = pd.concat([df_train, df_test], sort=False)

# Imputación de Age con la mediana
df_all["Age"] = df_all["Age"].fillna(df_all["Age"].median())

# Imputación de Embarked con el valor más frecuente
df_all["Embarked"] = df_all["Embarked"].fillna(df_all["Embarked"].mode()[0])

# Imputación de Fare faltante (en test.csv)
df_all["Fare"] = df_all["Fare"].fillna(df_all["Fare"].median())

# Nueva variable: tiene cabina (1/0)
df_all["Has_Cabin"] = df_all["Cabin"].notnull().astype(int)

# Codificación de variables categóricas
df_all = pd.get_dummies(df_all, columns=["Sex", "Embarked"], drop_first=True)

# Selección de variables relevantes
features = ['Pclass', 'Age', 'Fare', 'SibSp', 'Parch', 'Has_Cabin', 'Sex_male', 'Embarked_Q', 'Embarked_S']
df_all = df_all[["Survived", "is_train", "PassengerId"] + features]

# Separar de nuevo train y test
df_train_clean = df_all[df_all["is_train"] == 1].drop(columns=["is_train"])
df_test_clean = df_all[df_all["is_train"] == 0].drop(columns=["is_train", "Survived"])

# Separar X e y
X = df_train_clean[features]
y = df_train_clean["Survived"].astype(int)
X_test_final = df_test_clean[features]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

# RANDOM FOREST
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X, y)
rf_scores = cross_val_score(rf_model, X, y, cv=5, scoring='accuracy')
print("Random Forest Accuracy (CV 5 folds):", np.round(rf_scores.mean(), 4))

# LOGISTIC REGRESSION
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X, y)
log_scores = cross_val_score(log_model, X, y, cv=5, scoring='accuracy')
print("Logistic Regression Accuracy (CV 5 folds):", np.round(log_scores.mean(), 4))

# PREDICCIONES (usaremos el modelo Random Forest para el submission)
y_pred_rf = rf_model.predict(X_test_final)
```

```python
# Cargar PassengerId desde antes
submission = pd.DataFrame({
    "PassengerId": passenger_ids_test,
    "Survived": y_pred_rf.astype(int)
})

# Guardar archivo para Kaggle
submission_filename = f"submission_titanic_rf.csv"
submission.to_csv(submission_filename, index=False)
print(f"Archivo '{submission_filename}' generado correctamente ✅")
```

```python
models = ["Random Forest", "Logistic Regression"]
scores = [rf_scores.mean(), log_scores.mean()]

plt.bar(models, scores, color=["#1f77b4", "#ff7f0e"])
plt.ylim(0.7, 1.0)
plt.title("Comparación de Accuracy (CV 5 folds)")
plt.ylabel("Accuracy promedio")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

Se implementaron dos modelos clásicos de clasificación supervisada: Random Forest y Regresión Logística, ambos evaluados mediante validación cruzada de 5 pliegues (CV=5). Los resultados fueron los siguientes:

Random Forest alcanzó un accuracy promedio de 80.6%, destacándose como el mejor modelo en esta comparación.

Regresión Logística obtuvo un accuracy de 79.2%, ligeramente inferior.

Por lo tanto, se eligió Random Forest como modelo final para generar el archivo de predicción
